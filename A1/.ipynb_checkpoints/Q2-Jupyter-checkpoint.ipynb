```python
def predictedY(x, w):
    m = np.zeros((len(x), 2)) # initializing matrix
    
    powerList = [0, 1]
    
    for row in range(len(x)):
        m[row] = x[row]
        m[row] = np.power(m[row], powerList)
    
    y = np.dot(m, np.array(w))
    return y

def updateWeights(x, w, y, a):
    pred_y = predictedY(x, w)
    w[0] = w[0] - a*(pred_y-y)
    w[1] = w[1] - a*np.dot((pred_y-y), x)
    return w

def epoch(x, y, w, a):
    for i in range(len(x)):
        w = updateWeights(x, w, y, a)
    return w
```

```python
w = epoch(input_train, output_train, [3,5], 1e-6)
print(w)
```
